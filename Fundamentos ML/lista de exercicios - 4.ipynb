```python
import pandas as pd

file_path = '/content/drive/MyDrive/CÃ³pia de house-price.txt'

df = pd.read_csv(file_path, names=['x1', 'x2', 'y'])
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

X = df[['x1', 'x2']]
y = df['y']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

n=[10,20,25,30,40]

# Create a single figure with a 2x5 subplot grid
# fig, axes = plt.subplots(nrows, ncols, figsize)
fig, axes = plt.subplots(2, 5, figsize=(25, 10))

for idx, num in enumerate(n):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=num/100, random_state=42)

    # Plot for feature x1 (top row)
    sns.scatterplot(x=X_train['x1'], y=y_train, color='blue', label='Training Data', ax=axes[0, idx])
    sns.scatterplot(x=X_test['x1'], y=y_test, color='red', label='Testing Data', ax=axes[0, idx])
    axes[0, idx].set_title(f'Feature x1 ({100-num}/{num} Split)')
    axes[0, idx].set_xlabel('x1')
    axes[0, idx].set_ylabel('y')
    axes[0, idx].legend()
    axes[0, idx].grid(True)

    # Plot for feature x2 (bottom row)
    sns.scatterplot(x=X_train['x2'], y=y_train, color='blue', label='Training Data', ax=axes[1, idx])
    sns.scatterplot(x=X_test['x2'], y=y_test, color='red', label='Testing Data', ax=axes[1, idx])
    axes[1, idx].set_title(f'Feature x2 ({100-num}/{num} Split)')
    axes[1, idx].set_xlabel('x2')
    axes[1, idx].set_ylabel('y')
    axes[1, idx].legend()
    axes[1, idx].grid(True)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
